```python
#URL
driver.get('https://www.naukri.com/')

#Searching Parameter
search_job = driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys("Data Analyst")
search_loc = driver.find_element_by_xpath("//input[@id='qsb-location-sugg']")
search_loc.send_keys("Bangalore")

# Search Button
search_btn = driver.find_element_by_xpath("//div[@class='search-btn']/button")
search_btn.click()

job_titles = []
locations_list = []
company_names = []
experience_list = []

url = 'https://www.naukri.com/data-analyst-jobs-in-bangalore?k=data%20analyst&l=bangalore'

driver.get(url)

# Titles
titles_tags = driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
for i in titles_tags:
    title = i.text    
    job_titles.append(title)
#Location
location_tags = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
for i in location_tags:
    loc = i.text
    locations_list.append(loc)
# Company Names
companies_tags = driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
for i in companies_tags:
    comp_name = i.text
    company_names.append(comp_name)
# Experience
experience_tags = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
for i in experience_tags:
    exp = i.text
    experience_list.append(exp)
# Prepare Data Frame
dt = pd.DataFrame({'Title': job_titles[0:10], 'Location':locations_list[0:10], 'CompanyName': company_names[0:10], 'Experience':experience_list[0:10]})
dt
```

```python
#creating empty list
rating_list = []
review_list = []
full_review_list = []

#Create Function to scrap data
def scrapData(driver):
    #Scrap Rating
    rating_tag = driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']")
    for i in rating_tag:
        rating_list.append(i.text)
    rating_tag = driver.find_elements_by_xpath("//div[@class='_3LWZlK _1rdVr6 _1BLPMq']")
    for i in rating_tag:
        rating_list.append(i.text)
    #Scrap Review Summary
    summary_tag = driver.find_elements_by_xpath("//p[@class='_2-N8zT']")
    for i in summary_tag:
        review_list.append(i.text)
    #Scrap Full Review
    fullReview_tag = driver.find_elements_by_xpath("//div[@class='t-ZTKy']")
    for i in fullReview_tag:
        full_review_list.append(i.text)
```

```python
#creating empty list
company_list = []
salary_rcord_list = []
average_list = []
minimum_list = []
maximum_list = []
experience_list = []


#Scrap Company Name And Salary Records
company_data = driver.find_elements_by_xpath('//div[@class="name"]')
for i in company_data:
    data = i.text.split('\n')
    company_list.append(data[0])    
    salary_rcord_list.append(data[1])
    
#Scrap experience required    
experience_data = driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')
for i in experience_data:
    data = i.text.split('\n')
    experience_list.append(data[2])

#average salary, minimum salary, maximum salary    
salary_data = driver.find_elements_by_xpath('//div[@class="salary-range-wrapper"]')
for i in salary_data:
    data = i.text.split('\n')
    average_list.append(data[0])
    minimum_list.append(data[1])
    maximum_list.append(data[2])

# Prepare Data Frame
dt = pd.DataFrame({'Company Name': company_list[0:10], 'Salary Record': salary_rcord_list[0:10], 'Average Salary':average_list, 'Minimum Salary':minimum_list[0:10], 'Maximum Salary':maximum_list, 'Experience Require':experience_list})
dt
```
