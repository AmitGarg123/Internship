```python
def getIMDBData(url):
    names = []
    ratings = []
    years = []
    page = requests.get(url)#https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt
    soup = BeautifulSoup(page.content)
    for i in range(0, 2):
        if(i == 1):
            page = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt')
            soup = BeautifulSoup(page.content)
                    
        for i in soup.find_all('div', class_="lister-item-content"):
            no = i.find('span', class_="lister-item-index unbold text-primary")
            names.append(i.find('a').text)
            years.append(i.find('span', class_="lister-item-year text-muted unbold").text)
            ratings.append(i.find('strong').text)
    return pd.DataFrame({'Name': names, 'Rating':ratings, 'Year': years})
```

```python
getIMDBData('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
```

```python
def getODITeam(url):
    names = []
    matches = []
    points = []
    ratings = []
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    for index, i in enumerate(soup.find_all('tr')):
        if(index > 0 and index <= 10):
            row = i.find_all('td')
            names.append(row[1].find_all('span')[1].text)
            matches.append(row[2].text)
            points.append(row[3].text)
            ratings.append(row[4].text.strip())
    return pd.DataFrame({'Name': names, 'Matches':matches, 'Points': points, 'Rating':ratings})
```

```python
getODITeam('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
def getODIBatsmen(url):
    names = []
    teams = []    
    ratings = []
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    data = soup.find('div', class_="rankings-block__container")
    
    names.append(data.find('div', class_="rankings-block__banner--name").text)
    t = data.find('div', class_="rankings-block__banner--nationality").text.strip().split('\n')
    teams.append(t[0])
    ratings.append(t[1])
    for index, i in enumerate(data.find_all('tr')):
        if(index > 0):
            row = i.find_all('td')
            names.append(row[1].text.strip())
            teams.append(row[2].text.strip())
            ratings.append(row[3].text.strip())
    return pd.DataFrame({'Name': names, 'Team':teams, 'Rating':ratings})
```

```python
getODIBatsmen('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
```

```python
def getAllMobiles(url):
    names = []
    prices = []
    images = []
    ratings = []
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    for i in soup.find_all('div', class_="s-include-content-margin s-latency-cf-section s-border-bottom"):
        names.append(i.find('h2').text)
        prices.append(i.find('span', class_="a-offscreen").text)
        data = i.find('img', class_="s-image")
        images.append(data['src'])
        ratings.append(i.find('span', class_="a-icon-alt").text)        
    return pd.DataFrame({'Name': names, 'Price':prices, 'Image URL':images, 'Rating':ratings})
    
```

```python
def getHouseDetails(url):
    titles = []
    locations = []
    area = []
    emi = []
    prices = []
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    for i in soup.find_all('div', class_="nb__2JHKO"):
        titles.append(i.find('h2').text)
        locations.append(i.find('div', class_="nb__2CMjv").text)
        area.append(i.find('div', class_="nb__3oNyC").text)
        emi.append(i.find('div', {'id':"roomType"}).text)
        prices.append(i.find('div', {'id':"minDeposit"}).text)
        
    return pd.DataFrame({'Title': titles, 'Location':locations, 'Area':area, 'EMI':emi, 'Price':prices})
```

```python
def getRestaurantData(url):
    names = []
    cuisines = []
    locations = []
    ratings = []
    images = []
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    for i in soup.find_all('div', class_="restnt-card restaurant"):
        names.append(i.find('div', class_="restnt-info cursor").find('a').text)
        cuisines.append(i.find('span', class_='double-line-ellipsis').text.split('|')[1])
        locations.append(i.find('div', class_="restnt-loc ellipsis").text)
        ratings.append(i.find('div', {'class' : 'restnt-rating'}).text)
        images.append(i.find('img', class_='no-img')['data-src'])       
    
    return pd.DataFrame({'Name': names, 'Cuisine':cuisines, 'Location': locations, 'Rating':ratings, 'Image_URL':images})
```

```python
getRestaurantData('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
def getWeatherDetail(url):
    hour = []
    temperature = []
    wind = []
    weather = []
    humidity = []
    pressure = []
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    
    for index, i in enumerate(soup.find_all('tr')):
        if(index > 5):
            row = i.find_all('td')
            if(len(row) == 6):
                hour.append(row[0].text)
                temperature.append(row[2].text)
                wind.append(row[3].text)
                weather.append(row[1].text)
                humidity.append(row[4].text)
                pressure.append(row[5].text)
            
    return pd.DataFrame({'Hour':hour, 'Temperature':temperature, 'Wind':wind, 'Weather':weather, 'Humidity':humidity, 'Pressure':pressure})
```

```python
getWeatherDetail('https://en.tutiempo.net/delhi.html?data=last-24-hours')
```

```python
def getMonument(url):
    name = []
    description = []
    image = []
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    content = soup.find('div', class_="blog--single__content column--3-4 u-spacing-third")
    l = len(content.find_all('p'))
    for index, i in enumerate(content.find_all('p')):
        if(index < l-2 and i.find('strong') != None):
            name.append(i.text)
            nxt = i.find_next('p')
            description.append(nxt.text)
            image.append(nxt.find_next('p').find('img')['data-srcset'].split(' ')[0])
            
    return pd.DataFrame({'Name' : name, 'Description' : description, 'image' : image})
```

```python
getMonument('https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/')
```
